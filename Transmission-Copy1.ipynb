```python

def transmission(epsilon, huckel, coupling=-0.5):
    gamma_left = np.zeros(huckel.shape, order='F')
    gamma_left[0, 0] = coupling
    gamma_right = np.zeros(huckel.shape, order='F')
    gamma_right[-1, -1] = coupling

    temp_epsilon = np.zeros(huckel.shape)
    np.fill_diagonal(temp_epsilon, epsilon)
    g_ret = np.linalg.inv(
        temp_epsilon - huckel + (1j/2)*gamma_left + (1j/2)*gamma_right
    )

    g_adv = np.conjugate(g_ret)
    result = np.trace(
        np.matmul(np.matmul(np.matmul(gamma_left, g_ret), gamma_right), g_adv)
    )

    return result


def main():
    alpha = 0
    beta = -3

    # benzene = np.array([[alpha, beta, 0, 0, 0, beta],
    #                     [beta, alpha, beta, 0, 0, 0],
    #                     [0, beta, alpha, beta, 0, 0],
    #                     [0, 0, beta, alpha, beta, 0],
    #                     [0, 0, 0, beta, alpha, beta],
    #                     [beta, 0, 0, 0, beta, alpha]
    #                     ])
    linear_2 = np.array(
        [
            [alpha, beta],
            [beta, alpha], 
    )
    linear_4 = np.array(
        [
            [alpha, beta, 0, 0],
            [beta, alpha, beta, 0],
            [0, beta, alpha, beta],
            [0, 0, beta, alpha],
        ]
    )
    b1 = -3
    b2 = -4
    bithiophene = np.array(
        [
            [alpha, b1, 0, 0, b2, 0, 0, 0, 0],
            [b1, alpha, b1, 0, 0, 0, 0, 0, 0],
            [0, b1, alpha, b1, 0, 0, 0, 0, 0],
            [0, 0, b1, alpha, b2, 0, 0, 0, 0],
            [b2, 0, 0, b2, alpha, b2, 0, 0, b2],
            [0, 0, 0, 0, b2, alpha, b1, 0, 0],
            [0, 0, 0, 0, 0, b1, alpha, b1, 0],
            [0, 0, 0, 0, 0, 0, b1, alpha, b1],
            [0, 0, 0, 0, b2, 0, 0, b1, alpha]
        ]
    )
    db_octene = np.array(
        [
            [alpha, beta, 0, 0, 0, 0, 0, 0, beta],
            [beta, alpha, beta, 0, 0, 0, 0, 0, 0],
            [0, beta, alpha, beta, 0, 0, beta, 0, 0],
            [0, 0, b1, alpha, b2, 0, 0, 0, 0],
            [b2, 0, 0, b2, alpha, b2, 0, 0, b2],
            [0, 0, 0, 0, b2, alpha, b1, 0, 0],
            [0, 0, 0, 0, 0, b1, alpha, b1, 0],
            [0, 0, 0, 0, 0, 0, b1, alpha, b1],
            [0, 0, 0, 0, b2, 0, 0, b1, alpha]
        ]
    )

    grid = np.linspace(-6, 6, 2000)
    start = time()
    trans = [transmission(en, bithiophene) for en in grid]
    trans = [transmission(en, linear_4) for en in grid]
    plt.semilogy(grid, trans)
    trans = [transmission(en, linear_2) for en in grid]
    plt.semilogy(grid, trans)
    plt.show()


if __name__ == '__main__':
    exit(main())
```
